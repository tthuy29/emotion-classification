```
import re
import numpy as np
from nltk.tokenize import word_tokenize
import nltk
nltk.download('punkt')
def read():
    f1 = open('drive/colab/emotion_vn/x_tr_vn_pos.txt','r', encoding = 'utf-8')
    feature = []
    for l in f1:
          l = l.lstrip(' ')
          feature.append(l)
    f1.close()

    f2 = open('drive/colab/emotion_vn/pattern_vn.txt','r', encoding = 'utf-8')
    a = []
    for l in f2:
        l = l.replace('\n','')
        l = l.split('/')
        if l[1] not in a:
          a.append(l[1])
    f2.close()
    x_tr_data = []

    for i in feature:
          b = []
          for j in i.split():
              k = j.split('/')
              if (k[1] in a):
                    b.append(k[0])
              else:
                    if (len(k)==2):
                         b.append('')
          x_tr_data.append(b)
    
    f1 = open('drive/colab/emotion_vn/x_tst_vn_pos.txt','r', encoding = 'utf-8')
    feature = []
    for l in f1:
          l = l.lstrip(' ')
          feature.append(l)
    f1.close()

    f2 = open('drive/colab/emotion_vn/pattern_vn.txt','r', encoding = 'utf-8')
    a = []
    for l in f2:
        l = l.replace('\n','')
        l = l.split('/')
        if l[1] not in a:
          a.append(l[1])
    f2.close()
    x_tst_data = []

    for i in feature:
          b = []
          for j in i.split():
              k = j.split('/')
              if (k[1] in a):
                    b.append(k[0])
              else:
                    if (len(k)==2):
                         b.append('')
          x_tst_data.append(b)
    feature = []
 
    max_len = 25
    vecto_len = 300
    file_name1 = "drive/colab/emotion_vn/vectors.txt"
    dict1 = {}
    with open(file_name1, encoding = 'utf-8', mode = 'r') as f1:
        for line in f1:
            vecto = []
            value = line.split()
            word = value[0]
            vecto = np.asarray(value[1:], dtype = 'float32')
            dict1[word] = vecto
    f1.close()
   
    x_train = []
    for i in x_tr_data:
        j = []
        for k in i:
            k = k.lower()
            j.append(k)
        x_train.append(j)

    f2 = open('drive/colab/emotion_vn/x_train_fea_300_vn.txt','w', encoding='utf-8')
    for sen in x_train:
        l = 0
        for word in sen:
            if word in dict1:
                for p in range(vecto_len):
                    f2.write(str(dict1[word][p]))
                    f2.write(' ')
            else:
                for p in range(vecto_len):
                    f2.write(str(0))
                    f2.write(' ')
            l = l+1
            f2.write('\n')
            if l == max_len: break
        while l<max_len:
            for p in range(vecto_len):
                f2.write(str(0))
                f2.write(' ')
            f2.write('\n')
            l = l+1

    f2.close()
 
    x_train = []
    x_test = []
    for i in x_tst_data:
        j = []
        for k in i:
            k = k.lower()
            j.append(k)
        x_test.append(j)

    f3 = open('drive/colab/emotion_vn/x_test_fea_300_vn.txt','w', encoding='utf-8')
    for sen in x_test:
        l = 0
        for word in sen:
            if word in dict1:
                for p in range(vecto_len):
                    f3.write(str(dict1[word][p]))
                    f3.write(' ')
            else:
                for p in range(vecto_len):
                    f3.write(str(0))
                    f3.write(' ')
            l = l+1
            f3.write('\n')
            if l == max_len: break
        while l<max_len:
            for p in range(vecto_len):
                f3.write(str(0))
                f3.write(' ')
            f3.write('\n')
            l = l+1
    f3.close()

read()
```
